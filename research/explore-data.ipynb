# Data visualization

```python
import matplotlib.pyplot as plt
concat_df.head()
```

```python
def draw_hist_categorical_cols(df,cols):
    for col in cols:
        plt.figure(figsize=(9, 3))   

        plt.subplot(131)
        plt.xlabel(col)
        plt.hist(df[col]) #histogram

        plt.subplot(133)
        plt.xlabel(col)
        plt.scatter(concat_df[col].unique(),concat_df[col].value_counts()) #scatter plot
        
        plt.show()

```

```python
draw_hist_categorical_cols(concat_df,categorical_cols)
```
